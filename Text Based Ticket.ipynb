```python
import json

class TicketingSystem:
    def __init__(self, categories_file):
        self.categories = self.load_categories(categories_file)
        self.tickets = []
        self.ticket_id = 1

    def load_categories(self, categories_file):
        with open(categories_file, 'r') as file:
            data = json.load(file)
        return data["categories"]

    def display_categories(self):
        display_text = "Issue Categories:\n"
        for idx, category in enumerate(self.categories.keys(), 1):
            display_text += f"{idx}. {category}\n"
        print(display_text)
        return list(self.categories.keys())

    def display_subcategories(self, category):
        if category in self.categories:
            subcategories = self.categories[category]
            display_text = f"{category} Subcategories:\n"
            for idx, subcategory in enumerate(subcategories, 1):
                display_text += f"{idx}. {subcategory}\n"
            print(display_text)
            return subcategories
        else:
            print("Invalid category.")
            return []

    def create_ticket(self, category_idx, subcategory_idx, description):
        category = list(self.categories.keys())[category_idx]
        subcategory = self.categories[category][subcategory_idx]
        
        ticket = {
            "id": self.ticket_id,
            "category": category,
            "subcategory": subcategory,
            "description": description
        }
        self.tickets.append(ticket)
        print(f"Ticket {self.ticket_id} created successfully.")
        self.ticket_id += 1

    def list_tickets(self):
        if not self.tickets:
            print("No tickets created yet.")
        else:
            for ticket in self.tickets:
                print(f"ID: {ticket['id']}, Category: {ticket['category']}, Subcategory: {ticket['subcategory']}, Description: {ticket['description']}")

# Initialize the system
ticketing_system = TicketingSystem('categories.json')

# Display Categories
categories = ticketing_system.display_categories()

# Select Category
category_idx = int(input("Select a category (number): ")) - 1
subcategories = ticketing_system.display_subcategories(categories[category_idx])

# Select Subcategory
subcategory_idx = int(input(f"Select a subcategory (number) for {categories[category_idx]}: ")) - 1

# Describe the Issue
description = input("Describe the issue: ")
ticketing_system.create_ticket(category_idx, subcategory_idx, description)

# List Tickets
ticketing_system.list_tickets()
```
